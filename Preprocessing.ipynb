```python
import pandas as pd
import numpy as np
import re
```

```python
data = pd.read_csv('data.csv')
data.shape
```

```python
data
```

```python
data = data[data['Назначение ОРО'].apply(lambda x: isinstance(x, str))]
```

```python
def processing(st):
    st = st.lower()
    st = re.sub(r'[^\w\sа-яА-Я]', '', st)
    st = st.split()[0][:3]
    
    if st == 'зах':
        st = 'Захоронение'
    else:
        st = 'Хранение'
    return st
```

```python
data['Назначение ОРО'] = data['Назначение ОРО'].apply(lambda x: processing(x))
```

```python
a = data.groupby(['Наименование субъекта', 'Назначение ОРО']).count()
```

```python
a['Наименование субъекта'].unique().shape
```

```python
names = set(data['Наименование субъекта'].unique()[1:])
```

```python
len(names)
```

```python
# reset index to convert groupby results to DataFrame
a = a.reset_index()
```

```python
# rename count column to a more descriptive name
a = a.rename(columns={'№ ОРО в ГРОРО': 'Количество OPO'})
```

```python
a = a.iloc[:,:3]
```

```python
a['Наименование субъекта'].unique().shape
```

```python

# merge with original dataset using relevant columns as keys
data = data.merge(a, on=['Наименование субъекта', 'Назначение ОРО'], how='left')
```

```python
data['Наименование субъекта'].unique().shape
```

```python
data_small = data[['Наименование субъекта', 'Назначение ОРО', 'Количество OPO']]
```

```python
data_small = data_small.drop_duplicates().dropna()
```
